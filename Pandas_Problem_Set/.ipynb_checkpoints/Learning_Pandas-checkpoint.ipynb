### 3. Write a Python program to add, subtract, multiple and divide two Pandas Series.
#### Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]

```python
series1 = pd.Series([2, 4, 6, 8, 10])
series2 = pd.Series([1, 3, 5, 7, 9])

print("Addition: ")
print(series1 + series2)

print("Substraction: ")
print(series1 - series2)

print("Multiplication: ")
print(series2 * series1)

print("Division :")
print(series2 / series1)
```

#### 4. Write a Python program to get the powers of an array values element-wise.

- Note: First array elements raised to powers from second array
- Expected Output:
- Original array [0 1 2 3 4 5 6]

- First array elements raised to powers from second array, element-wise:
[ 0 1 8 27 64 125 216]       

```python
series1 ** series2
```

### 16. Write a Python program to calculate the mean score for each different student in DataFrame

```python
result = df.groupby('name')['score'].mean()
result = result.reset_index()
result.rename(columns={'score': 'mean_score'}, inplace=True)
print(result)
```
