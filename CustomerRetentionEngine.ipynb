```python
class DataLoader:
    def __init__(self, filepath):
        self.__df_raw = pd.read_csv(filepath)

    @property
    def Dataset(self):
        return self.__df_raw.copy()
```

```python
class CustomerSegmentationManager:
    @staticmethod
    def apply_segmentation(df):
        df['value_segment'] = pd.cut(
            df['total_spent'],
            bins = [0, 5000, 9000, np.inf],
            labels = ['Low', 'Medium', 'High']
        )

        df['activity_segment'] = pd.cut(
            df['last_purchase_days'],
            bins = [-1, 30, 90, np.inf],
            labels = ['Active', 'At Risk', 'Churned']
        )

        return df
```

```python
class AnalyticEngine:
    @staticmethod
    def correlation_analysis(df):
        corr1 = np.corrcoef(df['total_spent'],df['total_orders'])[0,1]
        corr2 = np.corrcoef(df['last_purchase_days'],df['total_orders'])[0,1]

        print(f"Correlation (Spent vs Orders): {corr1:.4f}")
        print(f"Correlation (Recency vs Orders): {corr2:.4f}")

    @staticmethod
    def city_analysis(df):
        result = df.groupby('city')['total_spent'].mean()
        print("\nAverage Spending by City:\n", result)
```
