Scenario 2
BigMart Product Pricing and Profitability Optimization Engine
Business Context
BigMart wants to review its pricing strategy to improve profitability without negatively impacting demand. Management needs insights into which products are mispriced relative to their sales performance.
Problem Statement
Build a PricingOptimizationEngine that analyzes pricing efficiency and recommends strategic price adjustments.
Business Requirements
Categorize products into price bands:


Low


Medium


Premium



Analyze:


Correlation between MRP and sales


Sales performance across outlet sizes


Identify:


Premium products with poor sales performance


Low-MRP products with unusually high sales


Conduct a pricing experiment:


Decrease MRP for underperforming premium products


Increase MRP slightly for high-performing low-price products


Recalculate projected revenue


Technical Constraints
Use NumPy for:


Correlation computation


Revenue simulation


Use Pandas for:


Binning using cut


Group-based analytics


Pricing rules must be configurable.


Visualization Requirements
Matplotlib:


Histogram of product MRP distribution


Plotly:


Interactive scatter plot with:


X-axis: MRP


Y-axis: Sales


Color: Outlet type


Hover: Product identifier and outlet identifier


OOP Constraints
Implement encapsulation for pricing logic


Use static methods to define price bands


No global variables
